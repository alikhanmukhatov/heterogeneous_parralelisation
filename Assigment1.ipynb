```python
%%writefile task1.cpp

#include<bits/stdc++.h>

using namespace std;

int main() {
    const int N = 50000;
    int* arr = new int[N];

    srand(time(nullptr));

    for (int i = 0; i < N; i++) {
        arr[i] = rand() % 100 + 1;
    }

    long long sum = 0;
    for (int i = 0; i < N; i++) {
        sum += arr[i];
    }

    double average = static_cast<double>(sum) / N;
    cout << "Среднее значение: " << average << endl;

    delete[] arr;
    return 0;
}
```

## Task 2

```python
%%writefile task2.cpp

#include <iostream>
#include <vector>
#include <chrono>
#include <climits>
#include <cstdlib>

int main() {
    const int N = 1'000'000;
    std::vector<int> arr(N);

    // Заполнение массива случайными числами
    for (int i = 0; i < N; ++i) {
        arr[i] = rand(); // можно заменить на rand() % 100000
    }

    int minVal = INT_MAX;
    int maxVal = INT_MIN;

    // Засекаем время начала
    auto start = std::chrono::high_resolution_clock::now();

    // Последовательный поиск минимума и максимума
    for (int i = 0; i < N; ++i) {
        if (arr[i] < minVal) minVal = arr[i];
        if (arr[i] > maxVal) maxVal = arr[i];
    }

    // Засекаем время окончания
    auto end = std::chrono::high_resolution_clock::now();

    std::chrono::duration<double, std::milli> duration = end - start;

    std::cout << "Минимум: " << minVal << std::endl;
    std::cout << "Максимум: " << maxVal << std::endl;
    std::cout << "Время выполнения: " << duration.count() << " мс" << std::endl;

    return 0;
}
```

## Task 3

```python
%%writefile task3.cpp
#include <iostream>
#include <vector>
#include <chrono>
#include <climits>
#include <cstdlib>
#include <omp.h>

int main() {
    const int N = 1'000'000;
    std::vector<int> arr(N);

    // Заполнение массива случайными числами
    for (int i = 0; i < N; ++i) {
        arr[i] = rand();
    }

    // ------------------ Последовательный поиск ------------------
    int minValSeq = INT_MAX;
    int maxValSeq = INT_MIN;

    auto startSeq = std::chrono::high_resolution_clock::now();

    for (int i = 0; i < N; ++i) {
        if (arr[i] < minValSeq) minValSeq = arr[i];
        if (arr[i] > maxValSeq) maxValSeq = arr[i];
    }

    auto endSeq = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double, std::milli> durationSeq = endSeq - startSeq;

    std::cout << "Последовательный поиск:\n";
    std::cout << "Минимум: " << minValSeq << ", Максимум: " << maxValSeq << "\n";
    std::cout << "Время: " << durationSeq.count() << " мс\n\n";

    // ------------------ Параллельный поиск с OpenMP ------------------
    int minValPar = INT_MAX;
    int maxValPar = INT_MIN;

    auto startPar = std::chrono::high_resolution_clock::now();

    #pragma omp parallel for reduction(min:minValPar) reduction(max:maxValPar)
    for (int i = 0; i < N; ++i) {
        if (arr[i] < minValPar) minValPar = arr[i];
        if (arr[i] > maxValPar) maxValPar = arr[i];
    }

    auto endPar = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double, std::milli> durationPar = endPar - startPar;

    std::cout << "Параллельный поиск (OpenMP):\n";
    std::cout << "Минимум: " << minValPar << ", Максимум: " << maxValPar << "\n";
    std::cout << "Время: " << durationPar.count() << " мс\n";

    return 0;
}
```

```python
%%writefile task4.cpp
#include <iostream>
#include <vector>
#include <chrono>
#include <cstdlib>
#include <omp.h>

int main() {
    const int N = 5'000'000;
    std::vector<int> arr(N);

    // Заполняем массив случайными числами
    for (int i = 0; i < N; ++i) {
        arr[i] = rand() % 100; // числа от 0 до 99
    }

    // ---------------- Последовательное вычисление ----------------
    long long sumSeq = 0;

    auto startSeq = std::chrono::high_resolution_clock::now();

    for (int i = 0; i < N; ++i) {
        sumSeq += arr[i];
    }

    double avgSeq = static_cast<double>(sumSeq) / N;

    auto endSeq = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double, std::milli> durationSeq = endSeq - startSeq;

    std::cout << "Последовательное среднее: " << avgSeq << "\n";
    std::cout << "Время: " << durationSeq.count() << " мс\n\n";

    // ---------------- Параллельное вычисление с OpenMP ----------------
    long long sumPar = 0;

    auto startPar = std::chrono::high_resolution_clock::now();

    #pragma omp parallel for reduction(+:sumPar)
    for (int i = 0; i < N; ++i) {
        sumPar += arr[i];
    }

    double avgPar = static_cast<double>(sumPar) / N;

    auto endPar = std::chrono::high_resolution_clock::now();
    std::chrono::duration<double, std::milli> durationPar = endPar - startPar;

    std::cout << "Параллельное среднее (OpenMP): " << avgPar << "\n";
    std::cout << "Время: " << durationPar.count() << " мс\n";

    return 0;
}
```
